Introduction
------------

The UK Met Office publishes the average temperature in England since 1659 as a text file at the following URL:

http://www.metoffice.gov.uk/hadobs/hadcet/ssn_HadCET_mean.txt

We can use the ``urllib`` module in Python to load the data at this URL as follows:

```python
import urllib
url = urllib.urlopen('http://www.metoffice.gov.uk/hadobs/hadcet/ssn_HadCET_mean.txt')
lines = url.readlines()
```

```python
lines[0:25]
```

We can then use the ``loadtxt`` function in Numpy to turn the text strings into a multi-dimensional array: 

```python
import numpy as np
data = np.loadtxt(lines[13:])
```

```python
data.shape
```

```python
data
```

Finally, we can use slicing to convert the multi-dimensional array into easy-to-use vectors:

```python
year = data[:,0]
wintertemp = data[:,1]
springtemp = data[:,2]
summertemp = data[:,3]
falltemp = data[:,4]
```

```python
year[0:50]
```

We can now plot the summer temperature in each year:

```python
import matplotlib.pyplot as plt
plt.plot(year, summertemp)
```

Exercise 4
==========

Compute the mean summer temperature for each decade *since 1660*, and plot it on top of the annual data with appropriate labels.

*Solution*:

```python
decade = np.arange(1660, 2010, 10)
meantemp = []
for i in decade:
    keep = (year >= i) & (year < i+10)
    meantemp.append(np.mean(summertemp[keep]))

plt.plot(year, summertemp, linewidth=1, color='blue')
plt.plot(decade+5, meantemp, linewidth=5, color='red')

plt.title('Summer temperatures in England since 1660')
plt.xlabel('Year')
plt.ylabel('Temperature')
```
